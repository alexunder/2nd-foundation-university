```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

dataframe = pd.read_csv("homework-C.csv")
display(dataframe)

ctable = pd.crosstab(index = dataframe['Location Preference'], columns = dataframe['Department'])

display(ctable)

c, p, dof, expected = chi2_contingency(ctable)

print("The Calculated Value is " + str(c))
print("The P Value is " + str(p))
print("The degree of freedom is " + str(dof))
print("The Expected table is \n" + str(expected))
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

dataframe = pd.read_csv("homework-D.csv")
display(dataframe)
table = pd.pivot_table(dataframe, index=["OBSERVED"])
display(table)
ctable = table.iloc[:3, :3]
display(table.iloc[:3, :3])
c, p, dof, expected = chi2_contingency(ctable)

print("The Calculated Value is " + str(c))
print("The P Value is " + str(p))
print("The degree of freedom is " + str(dof))
print("The Expected table is \n" + str(expected))
```
